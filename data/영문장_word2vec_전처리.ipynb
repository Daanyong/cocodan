```python
import pandas as pd

data=list()


# Load the Excel file
file_path = ['/content/drive/MyDrive/GDSC/1_구어체(1).xlsx',
             '/content/drive/MyDrive/GDSC/1_구어체(2).xlsx',
             '/content/drive/MyDrive/GDSC/2_대화체.xlsx',
             '/content/drive/MyDrive/GDSC/3_문어체_뉴스(1)_200226.xlsx',
             '/content/drive/MyDrive/GDSC/3_문어체_뉴스(2).xlsx',
             '/content/drive/MyDrive/GDSC/3_문어체_뉴스(3).xlsx',
             '/content/drive/MyDrive/GDSC/3_문어체_뉴스(4).xlsx',
             '/content/drive/MyDrive/GDSC/4_문어체_한국문화.xlsx',
             '/content/drive/MyDrive/GDSC/5_문어체_조례.xlsx',
             '/content/drive/MyDrive/GDSC/6_문어체_지자체웹사이트.xlsx']

for path in file_path:
  data.append(pd.read_excel(path))

# Display the first few rows of the dataframe to understand its structure
data[1].head()
```

```python
# Assuming that 'data' is a list of pandas DataFrames
combined_data = pd.concat(data, ignore_index=True)

# Display the first few rows of the combined dataframe to verify
print(combined_data.shape)
```

```python
from gensim.models import Word2Vec
import nltk
from nltk.tokenize import word_tokenize

# Ensure that the punkt tokenizer is downloaded
nltk.download('punkt')

# Extract the English sentences from the dataframe
english_sentences = combined_data['번역문'].astype(str).tolist()

# Tokenize the sentences
tokenized_sentences = [word_tokenize(sentence) for sentence in english_sentences]

# Let's view the first few tokenized sentences
tokenized_sentences[:5]
```

```python
import random
random.shuffle(tokenized_sentences)
```

```python
tokenized_sentences[:5]
```

```python
# Initialize the Word2Vec model
model = Word2Vec(vector_size=100, window=5, min_count=1, workers=4)

# Build vocabulary from the tokenized sentences
model.build_vocab(tokenized_sentences)

# Train the Word2Vec model
model.train(tokenized_sentences, total_examples=model.corpus_count, epochs=10, compute_loss=True)

# Save the model to a file
model_save_path = '/content/drive/MyDrive/GDSC/word2vec_model.model'
model.save(model_save_path)

# Check the vector for a sample word
sample_vector = model.wv['experience'] if 'experience' in model.wv else None
print(sample_vector)
```

```python
model_result = model.wv.most_similar("data")
print(model_result)
```
